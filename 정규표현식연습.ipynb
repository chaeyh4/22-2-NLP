```python
import re
```

```python
# 텍스트 내부에 패턴 존재하는지 유무를 출력하는 함수, re.search가 메인

def text_match(patterns, text):
  if re.search(patterns, text):
    print('일치 항목 존재')
  else:
    print('일치 항목 없음')
```

연습

```python
text_match("ab?",'ac')
text_match("ab?","abc")
text_match("ab?","abbc")
```

```python
text_match("ab*",'ac')
text_match("ab*","abc")
text_match("ab*","abbc")
```

```python
text_match("ab+",'ac')
text_match("ab+","abc")
text_match("ab+","abbc")
```

```python
text_match("a|b","c")
text_match("a|b","ac")
text_match("a|b","bc")
```

```python
text_match("ab{2,}c",'abc')
text_match("ab{2,}c",'abbc')
text_match("ab{2,}c",'abbb')
```

```python
text_match("b{3}",'b')
text_match("b{3}",'bb')
text_match("b{3}",'bbb')
```

```python
text_match("ab{3,5}?","abbbbbc")
```

```python
text_match("^a.*b$",'ab')
text_match("^a.*b$",'aeeeeeec')
```

```python
text_match("\w+\S*?$",'Tuffy eats pie, Loki eats peas!')
```

```python
text='Tuffy eats pie, Loki eats peas!'
for match in re.finditer("\w+\S*?$",text):
  s=match.start()
  e=match.end()
  print('%d:%d에서 "%s"를 찾았습니다.'%(s,e,text[s:e]))
```

```python
text_match("[ab]",'ab')
text_match("[cd]",'ab')
text_match("[ab]",'ac')
```

```python
text_match("[^ab]",'ab')
text_match("[^ab]",'ad')
text_match("[^ab]",'cd')
```

```python
text_match("[a-d]",'ac')
text_match("[a-d]",'ef')
```

```python
char1=re.compile(r'[^a-zA-Z0-9.]')
print(not bool(char1.search("ABCDEFabcedf123450.")))
print(not bool(char1.search('*&%@#!}{')))
```
