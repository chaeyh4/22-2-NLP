```python
import re
```

```python
#       , re.search 

def text_match(patterns, text):
  if re.search(patterns, text):
    print('  ')
  else:
    print('  ')
```

```python
text_match("ab?",'ac')
text_match("ab?","abc")
text_match("ab?","abbc")
```

```python
text_match("ab*",'ac')
text_match("ab*","abc")
text_match("ab*","abbc")
```

```python
text_match("ab+",'ac')
text_match("ab+","abc")
text_match("ab+","abbc")
```

```python
text_match("a|b","c")
text_match("a|b","ac")
text_match("a|b","bc")
```

```python
text_match("ab{2,}c",'abc')
text_match("ab{2,}c",'abbc')
text_match("ab{2,}c",'abbb')
```

```python
text_match("b{3}",'b')
text_match("b{3}",'bb')
text_match("b{3}",'bbb')
```

```python
text_match("ab{3,5}?","abbbbbc")
```

```python
text_match("^a.*b$",'ab')
text_match("^a.*b$",'aeeeeeec')
```

```python
text_match("\w+\S*?$",'Tuffy eats pie, Loki eats peas!')
```

```python
text='Tuffy eats pie, Loki eats peas!'
for match in re.finditer("\w+\S*?$",text):
  s=match.start()
  e=match.end()
  print('%d:%d "%s" .'%(s,e,text[s:e]))
```

```python
text_match("[ab]",'ab')
text_match("[cd]",'ab')
text_match("[ab]",'ac')
```

```python
text_match("[^ab]",'ab')
text_match("[^ab]",'ad')
text_match("[^ab]",'cd')
```

```python
text_match("[a-d]",'ac')
text_match("[a-d]",'ef')
```
